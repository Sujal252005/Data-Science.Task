```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('titanic.csv')

print(df1.head())
print(df1.info())
print(df1.describe())
print(df1.isnull().sum())

df1['age'] = df1['age'].fillna(df1['age'].median())
df1['embarked'] = df1['embarked'].fillna(df1['embarked'].mode()[0])
df1.drop(columns=['deck'], inplace=True)

print(df1['survived'].value_counts(normalize=True))
print(df1['pclass'].value_counts())

pclass_survival = df1.groupby('pclass')['survived'].mean()
plt.bar(pclass_survival.index.astype(str), pclass_survival.values, color='skyblue')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

plt.hist(df1['age'], bins=20, color='orange', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
