```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("income.csv")
print(df.head())

plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Income($)'], color='skyblue')

plt.title('Income by Category')
plt.xlabel('Name')
plt.ylabel('Income($)')
plt.xticks(rotation=45) 

for i in range(len(df)):
    plt.text(i, df['Income($)'][i] + 1000, str(df['Income($)'][i]), ha='center')

plt.tight_layout()
plt.show()
```
